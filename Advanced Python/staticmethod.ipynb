Sınıfın bir örneğine ya da sınıfın kendisine erişmeye ihtiyaç duymayan methodlardır.

Normal metotlar ilk parametre olarak self (nesneye erişim) veya cls (sınıfa erişim) alırken, statik metotlar hiçbir özel parametre almazlar. Onları bir sınıfın içine paketlenmiş, sınıfın genel amacıyla ilgili olan ama bağımsız çalışan "yardımcı fonksiyonlar" gibi düşünebiliriz.

### Temel Özellikler
- @staticmethod dekoratörü ile tanımlanırlar.

- Ne nesne verilerine (self), ne de sınıf verilerine (cls) ulaşabilirler.

- Hem sınıf üzerinden (Sınıf.metot()) hem de nesne üzerinden (nesne.metot()) çağrılabilirler.

```python
class Matematik:
    def __init__(self, isim):
        self.isim = isim

    def selamla(self):
        print(f"Merhaba, ben Robot {self.isim}.")

    @staticmethod
    def topla(sayi1, sayi2):
        return sayi1+sayi2
```

```python
# Kullanım:

# statik metoda ulaşmak için nesne oluşturmaya gerek yok.
sonuc = Matematik.topla(10,22)
print(sonuc)
```

```python
# Nesene üzerinden de çağrılabilir ama self'e erişemez.
robot = Matematik("Alex")
robot.selamla()
print(f"Toplama sonucu: {robot.topla(5, 4)}")
```

"self'e erişemez" demek, o metodun içinde bulunduğu nesnenin özel bilgilerini (değişkenlerini) görememesi ve değiştirememesi demektir.

Bir sınıfı "Araba Fabrikası", nesneyi ise o fabrikadan çıkan "Kırmızı bir Araba" gibi düşünürsek:

self kullanan metot: Arabanın içindeki şoför gibidir. Arabanın rengini bilir, yakıt miktarını kontrol eder.

static method: Arabanın torpidosunda duran bir hesap makinesi gibidir. Arabaya aittir, arabanın içindedir ama arabanın rengiyle veya motoruyla hiçbir bağı yoktur. Sadece ona verdiğin sayıları toplar.

```python
class Araba:
    def __init__(self, marka, renk):
        self.marka = marka  
        self.renk = renk    

    def bilgileri_goster(self):
        # Bu metot self'e erişebilir, çünkü self parametresi var
        print(f"Bu araba bir {self.marka} ve rengi {self.renk}.")

    @staticmethod
    def hiz_hesapla(yol, zaman):
        # DİKKAT: Burada 'self.marka' dersen hata alırsın!
        # Çünkü statik metot hangi nesnenin içinde olduğunu "bilmez" ve "ilgilenmez"
        return yol / zaman

# Kullanım:
benim_arabam = Araba("Tesla", "Kırmızı")

# 1. Normal metot self sayesinde nesne bilgilerini okur:
benim_arabam.bilgileri_goster() 

# 2. Statik metodu nesne üzerinden çağırsak bile:
sonuc = benim_arabam.hiz_hesapla(100, 2)
print(f"Hız: {sonuc}")
```

### Metot Türü - İlk Parametre - Erişim Alanı

Instance Method -- self -- Nesne özelliklerine ve sınıf özelliklerine erişir.

Class Method -- cls -- Sadece sınıf özelliklerine erişir.

Static Method -- Yok --  Hiçbirine erişemez, bağımsız çalışır.
