Python'da @property dekoratörü, bir sınıf metodunu bir nitelik (attribute) gibi çağırmamızı sağlar. Yani, arkada bir fonksiyon çalışmasına rağmen, biz sanki bir değişkene erişiyormuşuz gibi nesne.metod yerine nesne.nitelik şeklinde işlem yaparız.

Encapsulation

 ### Neden ve Ne Zaman Kullanılmalı?
Normalde sınıflarda verileri korumak için get_degisken() ve set_degisken() gibi metodlar yazarız. Ancak bu, kodu kalabalıklaştırır. @property şu durumlarda imdadımıza yetişir:

- Validation: Bir değişkene değer atanırken (örneğin yaşın negatif olmaması gibi) kontrol yapmak istediğinizde.

- Read-only Özellikler: Bir değerin dışarıdan değiştirilmesini istemediğinizde.

- Hesaplanmış Özellikler: Bir değer, diğer niteliklere bağlı olarak dinamik hesaplanıyorsa (örneğin ad ve soyaddan tam_ad oluşturmak).

- Geriye Dönük Uyumluluk: Başta halka açık (public) olan bir değişkeni, daha sonra kontrol altına almak istediğinizde kodun geri kalanını bozmadan metod haline getirebilirsiniz.

Üç temel bileşeni vardır: Getter, Setter ve Deleter 

```python
class Calisan:
    def __init__(self, ad, maas):
        self.ad = ad
        self._maas = maas   
        """
        _ işareti protected anlamı taşır.
        anlamı: dışarıdan kurcalama
        
        __ ise private anlamına gelir.
        anlamı: dışarıdan erişemezsin (nesne.__maas ile) Attribute Error verir.
        """

    
    @property
    def maas(self):
        # getter: maaşı okumaya çalıştıgımızda burası çalışır.
        if self._maas is None:
            return "Maaş bilgisi silinmiş veya girilmemiş."
        return f"{self._maas} TL" 
    
    
    @maas.setter
    def maas(self,yeni_maas):
        # setter. maaşı değiştirmeye çalıştığımızda burası çalışır.
        if yeni_maas < 0:
            print("Hata: Maaş negatif olamaz.")
        else:
            self._maas = yeni_maas

    @maas.deleter
    def maas(self):
        # del personel.maas dendiğinde çalışır.
        print(f"UYARI: {self.ad} isimli personelin maaş verisi siliniyor...")
        self._maas = None
```

```python
# kullanım senaryosu
personel = Calisan("Ahmetcan", 50_000)

print(personel.maas)  # getter çalışır
personel.maas = 100_000  # setter çalışır.
print(personel.maas) # getter çalışır
personel.maas= -250  # setter çalışır, hata verir.
```

```python
print(personel.maas)  # getter
del personel.maas  # deleter
print(personel.maas)
```

### Getter, Setter ve Deleter Mantığı
- Getter (@property): Metodu bir özellik gibi okumanızı sağlar. obj.ozellik dendiğinde tetiklenir.

- Setter (@isim.setter): Özelliğe değer atamanızı sağlar. obj.ozellik = değer dendiğinde tetiklenir.

- Deleter (@isim.deleter): Özellik silindiğinde çalışır. del obj.ozellik dendiğinde tetiklenir.
