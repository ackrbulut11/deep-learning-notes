```python
class Pizza:
    def __init__(self, malzemeler):
        self.malzemeler = malzemeler

    def __repr__(self):
        return f"Malzemeler: {','.join(self.malzemeler)}"

    @classmethod
    def margarita(cls):
        # Müşteri malzemeleri tek tek saymasın diye standart recete
        recete = ["mozarella", "domates"]
        return cls(recete) # Pizza(["mozarella", "domates"]) demekle aynı

    @classmethod
    def pepperoni(cls):
        recete = ["mozarella", "domates", "pepperoni"]
        return cls(recete)
```

```python
# Kullanım: 

# 1. Özel Sipariş (Normal Metot / __init__):
ozel = Pizza(["peynir", "mısır", "zeytin", "sucuk", "domates"])
print(f"Özel sipariş -- {ozel}")

# 2. Menüden Sipariş (Class Method / cls):
menuden = Pizza.margarita()
print(f"Menüden sipariş -- {menuden}")
```

#### Özetle cls ne yaptı?
- self olsaydı: "Bu pizzanın üzerine mısır ekle" (Var olan bir pizzayı değiştirirdi).

- cls olduğu için: "Bana mutfaktan standart bir Margarita çıkart" (Sıfırdan, belli bir kalıba göre nesne üretti).
