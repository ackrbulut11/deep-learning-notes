```python
from abc import ABC, abstractmethod
```

```python
class Animal(ABC):

    @abstractmethod
    def ses_cikar(self):
        # bu bir soyut metottur. gövdesi boştur
        # sesi alt sınıflar belirleyecek
        pass

    def uyu(self):
        # soyut olmayan normal metot
        print("Uyuyor... zzz...")

class Cat(Animal):
    def ses_cikar(self):
        return "Miyav"

class Dog(Animal):
    def ses_cikar(self):
        return "Hav hav"
    
```

```python
kedi = Cat()
print(kedi.ses_cikar())
```

```python
köpek = Dog()
print(köpek.ses_cikar())
```

```python
hayvan = Animal()
print(hayvan.ses_cikar())
```

Eşek diye bir sınıf eklediğini düşün. Eğer Hayvan sınıfını miras alıyorsan ve ses_cikar metodunu yazmayı unutursan, Python kodun çalışmasına izin vermez. Bu bir sözleşmedir. "Eğer bir Hayvan olacaksan, ses çıkarmayı bilmelisin!" der.
