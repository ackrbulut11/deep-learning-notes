### Manual Learning 

```python
from torch import nn
```

```python
class SimpleLinearRegressionModel(nn.Module): # inheriting neural network module
    def __init__(self):
        super().__init__()
        self.weights = nn.Parameter(torch.randn(1, dtype=torch.float)) # random weights
        self.bias = nn.Parameter(torch.randn(1, dtype=torch.float))  # random bias

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.weights * x + self.bias
```

```python
torch.manual_seed(31)

model = SimpleLinearRegressionModel()
model = torch.compile(model)  
```

```python
list(model.parameters())
```

```python
model.state_dict()
```

```python
import torch._dynamo
torch._dynamo.config.suppress_errors = True

# compiler hatası alıyordu aşagıdaki kod. onu engellemek için.
```

```python
# tahmin yapacagız. Başarısız olacak çünkü henüz eğitmedik.

with torch.inference_mode():   # turns off gradient tracking, it is not training.
    y_preds = model(x_test)
```

```python
len(x_test), len(y_preds)
```

### Optimize loss function

```python
mse = nn.MSELoss()
# mae = nn.F1Loss()

# create optimizer
optimizer = torch.optim.SGD(params = model.parameters(), lr = 0.01)
```

```python
torch.manual_seed(31)

epochs = 200
train_loss_values = []
test_loss_values = []
epoch_counts = []

for epoch in range(epochs):
    model.train()

    y_pred = model(x_train)  # training data using forward()
    loss = mse(y_pred, y_train) 
    optimizer.zero_grad()  # optimizer gradient numbers accumulate every epoch, we need to reset it
    loss.backward()
    optimizer.step()  # optimize weights

    with torch.inference_mode():
        test_pred = model(x_test)
        test_loss = mse(test_pred, y_test.type(torch.float))

        if epoch % 5 == 0:
            epoch_counts.append(epoch)
            test_loss_values.append(test_loss.detach().numpy())
            train_loss_values.append(loss.detach().numpy())
            print(f"Epoch: {epoch} |  Train Loss: {loss} | Test Loss: {test_loss} ")
```
