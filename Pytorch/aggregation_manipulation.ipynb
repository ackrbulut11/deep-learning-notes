```python
x.unsqueeze(dim=0)  # başa bir boyut ekledi. 
```

```python
x.unsqueeze(dim=1) 
```

- dim=0: (1, A, B) -> Veriyi bir "paket" içine koyar (Batch eklemek gibi).

- dim=1: (A, 1, B) -> Satırlar arasına bir derinlik ekler.

- dim=2: (A, B, 1) -> Her sütun verisini kendi içinde bir boyuta hapseder.

- dim=0 (En Başa Ekleme)
    Yeni boyutu en dışa ekler. Bu genellikle "Batch Size" (yığın boyutu) eklemek için kullanılır.

    İşlem: x.unsqueeze(0)

    Sonuç: (1, 3, 4)

    Anlamı: "Elimde 3x4'lük veriden oluşan 1 adet paket var."

    Görünüm: [ [ [...] ] ]

- dim=-1 (En Sona Ekleme)
    -1 her zaman "sonuncu" demektir. Yeni boyutu en içe ekler.

    İşlem: x.unsqueeze(-1)
    
    Sonuç: (3, 4, 1)

    Görünüm: [ [ [.], [.] ] ]
